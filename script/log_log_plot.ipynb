```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV data
data = pd.read_csv('../data/data_frequency_power.csv')

# Get a list of unique boards
boards = data['board'].unique()

# Create the plot
plt.figure(figsize=(10, 8))

# Plot data for each board, up to 10 boards
for i, board in enumerate(boards[:10]):
    board_data = data[data['board'] == board]
    plt.loglog(board_data['power'], board_data['freq'], marker='o', linestyle='-', label=f'Board {board}')

# Add titles and labels
plt.title('Log-Log Plot of Power vs Frequency for Different Boards')
plt.xlabel('Power (log scale)')
plt.ylabel('Frequency (log scale)')
plt.legend()

# Show grid
plt.grid(True, which="both", ls="--")

# Display the plot
plt.show()
```
